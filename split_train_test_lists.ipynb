# 训练集 / 测试集名单

根据现有目录只生成名单，不移动或复制文件。

- **测试集**：`dataset/lowres_4x_simple/original` 下的 434 张图片的文件名 → `dataset/test_list.txt`
- **训练集**：`dataset/highres/original` 中「不在测试集对应名单里」的 1047 张图片的文件名 → `dataset/train_list.txt`

匹配方式：按**文件名主干**（不含扩展名）判断是否为同一张图，避免 .png/.jpg 不一致导致漏判。

```python
import os

PROJECT_DIR = os.getcwd()
LOWRES_DIR = os.path.join(PROJECT_DIR, 'dataset', 'lowres_4x_simple', 'original')
HIGHRES_DIR = os.path.join(PROJECT_DIR, 'dataset', 'highres', 'original')
OUT_DIR = os.path.join(PROJECT_DIR, 'dataset')

os.makedirs(OUT_DIR, exist_ok=True)

# 测试集：lowres_4x_simple 下的全部文件名（434 张）
exts = ('.png', '.jpg', '.jpeg', '.PNG', '.JPG', '.JPEG')
test_files = []
for f in sorted(os.listdir(LOWRES_DIR)):
    if f.endswith(exts):
        test_files.append(f)

test_stems = {os.path.splitext(f)[0] for f in test_files}
print(f'测试集（来自 lowres_4x_simple）：{len(test_files)} 张')

# 训练集：highres 中「主干不在 test_stems 里」的文件（1047 张）
train_files = []
for f in sorted(os.listdir(HIGHRES_DIR)):
    if f.endswith(exts):
        stem = os.path.splitext(f)[0]
        if stem not in test_stems:
            train_files.append(f)

print(f'训练集（highres 中其余）：{len(train_files)} 张')
print(f'highres 总张数：{len(test_files) + len(train_files)}（应与 434+1047=1481 一致）')
```

```python
# 写入 dataset/test_list.txt（434 个名字，一行一个）
test_path = os.path.join(OUT_DIR, 'test_list.txt')
with open(test_path, 'w', encoding='utf-8') as f:
    f.write('\n'.join(test_files))
print(f'已写入: {test_path}')

# 写入 dataset/train_list.txt（1047 个名字，一行一个）
train_path = os.path.join(OUT_DIR, 'train_list.txt')
with open(train_path, 'w', encoding='utf-8') as f:
    f.write('\n'.join(train_files))
print(f'已写入: {train_path}')
```
